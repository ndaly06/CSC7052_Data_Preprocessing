```python
# imports the neccessary python modules
import requests
import lxml.html as lh
import numpy as np
import os
import pandas as pd

# scraping guide
#https://towardsdatascience.com/web-scraping-html-tables-with-python-c9baba21059

# obtains the machine username for cleaner file paths
user = os.environ["HOME"]
```

```python
# website url
url='https://www.acarsd.org/Fleetlist_EZY_Easyjet.html'
#Create a handle, page, to handle the contents of the website
page = requests.get(url)
#Store the contents of the website under doc
doc = lh.fromstring(page.content)
#Parse data that are stored between <tr>..</tr> of HTML
tr_elements = doc.xpath('//tr')
```

```python
tr_elements = doc.xpath('//tr')
#Create empty list
col=[]
i=0
#For each row, store each first element (header) and an empty list
for t in tr_elements:
    i+=1
    name=t.text_content()
    print('%d:"%s"'%(i,name))
    col.append((name,[]))
```

```python
# creates a dataframe for the planes data
planes_data = pd.DataFrame(col)

planes_data = planes_data[[0]]

planes_data.columns = ['Planes']

planes_data.head()
```

```python
# splits the single plane column into multiple columns 
planes_data = planes_data['Planes'].str.split(' ', expand=True)

#
planes_data = planes_data.iloc[:, [0, 2]]

# renames the columns
planes_data.columns = ['TailNumber', 'Model']

planes_data.head()
```

```python
# drops rows containing specific values
planes_data = planes_data[planes_data.Model != 'B737-73V']
planes_data = planes_data[planes_data.Model != 'Airbus']

planes_data.head()
```

```python
# replaces plane model values
planes_data = planes_data.replace('A319-111', 'A319-100')
planes_data = planes_data.replace('A320-214', 'A320-200')

planes_data.head()
```

```python
# obtains first 5 and last 5 rows from planes data
planes_data1 = planes_data[:5]  # same as df.head(10)
planes_data2 = planes_data[-5:]

planes_data = pd.concat([planes_data1, planes_data2])
```

```python
# outputs the planes data to csv
planes_data.to_csv(user+"/Documents/CSC7052_Data_Preprocessing/data/planes.csv", index=False)
```
